**Selected N-Queen Problem**

**N - Queens problem is to place n - queens in such a manner on an n x n chessboard that no queens attack each other by being in the same row, column or diagonal.**

It can be seen that for n =1, the problem has a trivial solution, and no solution exists for n =2 and n =3. So first we will consider the 4 queens problem and then generate it to n - queens problem.

```python
# Function to check if two queens threaten each other or not
def canPlace(board, r, c):
 
    # return false if two queens share the same column
    for i in range(r):
        if board[i][c] == 'Q':
            return False
 
    # return false if two queens share the same `` diagonal
    (i, j) = (r, c)
    while i >= 0 and j >= 0:
        if board[i][j] == 'Q':
            return False
        i = i - 1
        j = j - 1
 
    # return false if two queens share the same `/` diagonal
    (i, j) = (r, c)
    while i >= 0 and j < len(board):
        if board[i][j] == 'Q':
            return False
        i = i - 1
        j = j + 1
 
    return True
```

Above function Check if Queens are attacking or not!

```python
def printSolution(board):
    for r in board:
        print(str(r).replace(',', '').replace('\'', ''))
    print()
```

Above funtion Replaces comma and inverted coma from matrix representation.

```python
def nQueen(board, r):
 
    # if `N` queens are placed successfully, print the solution
    if r == len(board):
        printSolution(board)
        return
 
    # place queen at every square in the current row `r`
    # and recur for each valid movement
    for i in range(len(board)):
 
        # if no two queens threaten each other
        if canPlace(board, r, i):
            # place queen on the current square
            board[r][i] = 'Q'
 
            # recur for the next row
            nQueen(board, r + 1)
 
            # backtrack and remove the queen from the current square
            board[r][i] = '–'
```

Above fuction Checks if Queen can be placed

```python
if __name__ == '__main__':
 
    # `N × N` chessboard
    
    print("_____________________________________________\n")
    N = int(input("Specify board or chess dimensions : "))
 
    # `board[][]` keeps track of the position of queens in
    # the current configuration
    board = [['–' for x in range(N)] for y in range(N)]
 
    nQueen(board, 0)
```
